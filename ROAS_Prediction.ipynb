```python
# 1. Define the lists of countries for the top 3 Tiers
tier_1 = ['US', 'GB', 'AU', 'CA', 'NZ', 'IE']

tier_2 = [
    'JP', 'KR', 'TW', 'HK', 'SG', 'AE', 'SA', 'DE', 'FR', 'NL',
    'IT', 'ES', 'SE', 'NO', 'FI', 'DK', 'CH', 'AT', 'BE', 'LU', 'IL'
]

tier_3 = [
    'BR', 'MX', 'IN', 'ID', 'VN', 'TH', 'PH', 'TR', 'RU', 'ZA',
    'AR', 'CO', 'CL', 'PE', 'MY', 'EG', 'PK', 'BD', 'NG', 'KZ', 'UA'
]

# 2. Create the categorization function
def categorize_country(country_code):
    if pd.isna(country_code):
        return 'Unknown'
    elif country_code in tier_1:
        return 'Tier_1'
    elif country_code in tier_2:
        return 'Tier_2'
    elif country_code in tier_3:
        return 'Tier_3'
    else:
        return 'Tier_4'

# 3. Apply the function to create the new column in train_df_copy
train_df_copy['Country_Tier'] = train_df_copy['Country'].apply(categorize_country)
test_df_copy['Country_Tier'] = test_df_copy['Country'].apply(categorize_country)

train_df_copy = train_df_copy.drop('Country', axis=1)
test_df_copy = test_df_copy.drop('Country', axis=1)
```

```python
from sklearn.preprocessing import OrdinalEncoder

# ── 1. Country Tier → Ordinal (Tier_1=4, Tier_2=3, Tier_3=2, Tier_4=1) ──
train_df_copy['Country_Tier'] = train_df_copy['Country_Tier'].replace('Unknown', 'Tier_4')
test_df_copy['Country_Tier']  = test_df_copy['Country_Tier'].replace('Unknown', 'Tier_4')

tier_order  = ['Tier_4', 'Tier_3', 'Tier_2', 'Tier_1']
ordinal_enc = OrdinalEncoder(categories=[tier_order], dtype=int)

# Fit trên train, transform cả 2
train_df_copy['Country_tier_enc'] = ordinal_enc.fit_transform(train_df_copy[['Country_Tier']]) + 1
test_df_copy['Country_tier_enc']  = ordinal_enc.transform(test_df_copy[['Country_Tier']])      + 1
# Kết quả: Tier_1=4, Tier_2=3, Tier_3=2, Tier_4=1

# ── 2. Platform → Binary ──
train_df_copy['is_ios'] = (train_df_copy['Platform'] == 'iOS').astype(int)
test_df_copy['is_ios']  = (test_df_copy['Platform']  == 'iOS').astype(int)

# ── 3. One-Hot Encoding — fit trên train rồi align test ──
train_df_copy = pd.get_dummies(train_df_copy, columns=['Package_name'], prefix='game', dtype=int)
test_df_copy  = pd.get_dummies(test_df_copy,  columns=['Package_name'], prefix='game', dtype=int)

# Align test theo đúng cột của train — cột game mới (I→M) fill 0
game_cols    = [c for c in train_df_copy.columns if c.startswith('game_')]
test_df_copy = test_df_copy.reindex(columns=test_df_copy.columns.tolist()[:] , fill_value=0)
test_df_copy = test_df_copy.reindex(columns=train_df_copy.columns, fill_value=0)

# ── 4. Drop cột gốc không cần nữa ──
train_df_copy.drop(columns=['Country_Tier', 'Platform'], inplace=True)
test_df_copy.drop(columns=['Country_Tier', 'Platform'],  inplace=True, errors='ignore')
```

```python
def remove_outliers_iqr(df, cols, multiplier=1.5):
    df_clean = df.copy()
    total_removed = 0

    for col in cols:
        Q1  = df_clean[col].quantile(0.25)
        Q3  = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - multiplier * IQR
        upper = Q3 + multiplier * IQR

        mask    = (df_clean[col] >= lower) & (df_clean[col] <= upper)
        removed = (~mask).sum()
        total_removed += removed

        print(f"  {col:<22} Q1={Q1:.3f}  Q3={Q3:.3f}  IQR={IQR:.3f}"
              f"  upper={upper:.3f}  removed={removed:,}")

        df_clean = df_clean[mask]

    print(f"\n  Trước: {len(df):,}  →  Sau: {len(df_clean):,}"
          f"  (đã xóa {total_removed:,} rows, {100*total_removed/len(df):.1f}%)")
    return df_clean


# Chỉ apply trên Ad và Total — KHÔNG đụng IAP vì IQR=0
outlier_cols = ['D30_ROAS_Ad', 'D30_ROAS_Total']
train_df_copy = remove_outliers_iqr(train_df_copy, outlier_cols, multiplier=1.5)
```

```python
train_df_copy['log_Installs'] = np.log1p(train_df_copy['Installs'])
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, LogisticRegression
from sklearn.metrics import mean_absolute_error, r2_score, classification_report
X_full = train_df_copy.drop(columns=['D30_ROAS_Total', 'D30_ROAS_Ad', 'D30_ROAS_IAP'])
y_full = train_df_copy['D30_ROAS_Total']

# Split 80/20
X_train, X_val, y_train, y_val = train_test_split(
    X_full, y_full, test_size=0.2, random_state=42
)
```

```python
import numpy as np
from sklearn.linear_model import Ridge, LogisticRegression
from sklearn.metrics import mean_absolute_error, r2_score, classification_report

# CHUẨN BỊ X, y CHO TỪNG MODEL
game_cols = [c for c in X_train.columns if c.startswith('game_')]
feat_ad  = ['D0_ROAS_Ad', 'D1_ROAS_Ad', 'D2_ROAS_Ad', 'D3_ROAS_Ad',
            'log_Installs', 'D1_Ret'] + game_cols + ['Country_tier_enc']

feat_iap = ['D0_ROAS_IAP', 'D1_ROAS_IAP', 'D2_ROAS_IAP', 'D3_ROAS_IAP',
            'log_Installs', 'D1_Ret'] + game_cols + ['Country_tier_enc']

X_train_ad  = X_train[feat_ad]
X_train_iap = X_train[feat_iap]
X_val_ad    = X_val[feat_ad]
X_val_iap   = X_val[feat_iap]

y_train_ad  = train_df_copy.loc[X_train.index, 'D30_ROAS_Ad']
y_train_iap = train_df_copy.loc[X_train.index, 'D30_ROAS_IAP']
y_val_ad    = train_df_copy.loc[X_val.index,   'D30_ROAS_Ad']
y_val_iap   = train_df_copy.loc[X_val.index,   'D30_ROAS_IAP']

LOG_CLIP = np.log1p(y_train_ad.quantile(0.999))
# MODEL 1 — AD (Ridge trên log)
print("Fitting Model Ad...")
nz_ad  = y_train_ad > 0
reg_ad = Ridge(alpha=1.0)
reg_ad.fit(X_train_ad[nz_ad], np.log1p(y_train_ad[nz_ad]))
print(f"  R²(log-space train): {reg_ad.score(X_train_ad[nz_ad], np.log1p(y_train_ad[nz_ad])):.4f}")
# MODEL 2 — IAP (Zero-inflated)
print("\nFitting Model IAP...")

# Stage 1: Classify IAP = 0 hay > 0
clf_iap = LogisticRegression(max_iter=500, C=1.0, random_state=42)
clf_iap.fit(X_train_iap, (y_train_iap > 0).astype(int))

# Stage 2: Regress trên nonzero rows
nz_iap  = y_train_iap > 0
reg_iap = Ridge(alpha=1.0)
reg_iap.fit(X_train_iap[nz_iap], np.log1p(y_train_iap[nz_iap]))
print(f"  R²(log-space train): {reg_iap.score(X_train_iap[nz_iap], np.log1p(y_train_iap[nz_iap])):.4f}")

print(f"\n  Classifier IAP (Val):")
print(classification_report(
    (y_val_iap > 0).astype(int),
    clf_iap.predict(X_val_iap),
    target_names=['IAP=0', 'IAP>0'], digits=3
))
# PREDICT FUNCTIONS
def predict_ad(X):
    log_pred = np.clip(reg_ad.predict(X[feat_ad]), 0, LOG_CLIP)
    return np.maximum(np.expm1(log_pred), 0)

def predict_iap(X, threshold=0.5):
    prob     = clf_iap.predict_proba(X[feat_iap])[:, 1]
    log_pred = np.clip(reg_iap.predict(X[feat_iap]), 0, LOG_CLIP)
    val_iap  = np.maximum(np.expm1(log_pred), 0)
    return np.where(prob >= threshold, val_iap, 0.0)
# ĐÁNH GIÁ TRAIN VÀ VAL
def evaluate(yt, yp, label):
    mae  = mean_absolute_error(yt, yp)
    r2   = r2_score(yt, yp)
    mape = np.mean(np.abs((yt - yp) / np.where(yt == 0, 1, yt))) * 100
    print(f"  {label:<20} MAE={mae:.4f}   R²={r2:.4f}   MAPE={mape:.1f}%")

for split, X, y_ad, y_iap, y_tot in [
    ("Train", X_train, y_train_ad, y_train_iap, y_train),
    ("Val",   X_val,   y_val_ad,   y_val_iap,   y_val),
]:
    pa = predict_ad(X)
    pi = predict_iap(X)
    pt = pa + pi
    print(f"\n=== {split} ===")
    evaluate(y_ad.values,  pa, "D30_ROAS_Ad")
    evaluate(y_iap.values, pi, "D30_ROAS_IAP")
    evaluate(y_tot.values, pt, "D30_ROAS_Total")
```

```python
# Xem hệ số của Model Ad
coef_ad = pd.Series(reg_ad.coef_, index=feat_ad)
print("=== Model Ad — Coefficients ===")
print(f"  Intercept (β0): {reg_ad.intercept_:.4f}")
print()
print(coef_ad.sort_values(ascending=False).to_string())

print()

# Xem hệ số của Model IAP (regressor)
coef_iap = pd.Series(reg_iap.coef_, index=feat_iap)
print("=== Model IAP — Coefficients ===")
print(f"  Intercept (β0): {reg_iap.intercept_:.4f}")
print()
print(coef_iap.sort_values(ascending=False).to_string())
```

```python
# PREDICT DATASET B
# Đảm bảo test có đủ features
test_df_copy['log_Installs'] = np.log1p(test_df_copy['Installs'])

# Align columns theo X_train — game mới fill 0
test_df_copy = test_df_copy.reindex(columns=X_train.columns, fill_value=0)

# Predict
test_df_copy['D30_ROAS_Ad']    = predict_ad(test_df_copy)
test_df_copy['D30_ROAS_IAP']   = predict_iap(test_df_copy)
test_df_copy['D30_ROAS_Total'] = test_df_copy['D30_ROAS_Ad'] + test_df_copy['D30_ROAS_IAP']

print("=== Predictions Dataset B ===")
print(f"  D30_Total mean  : {test_df_copy['D30_ROAS_Total'].mean():.4f}")
print(f"  D30_Total median: {test_df_copy['D30_ROAS_Total'].median():.4f}")
print(f"  D30_Ad    mean  : {test_df_copy['D30_ROAS_Ad'].mean():.4f}")
print(f"  D30_IAP   mean  : {test_df_copy['D30_ROAS_IAP'].mean():.4f}")
print(f"  Zero %          : {100*(test_df_copy['D30_ROAS_Total']==0).mean():.1f}%")
print(f"\n{test_df_copy[['D30_ROAS_Ad','D30_ROAS_IAP','D30_ROAS_Total']].describe().round(4)}")
```
